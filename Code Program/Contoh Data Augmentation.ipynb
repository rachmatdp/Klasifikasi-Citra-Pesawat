```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import matplotlib.pyplot as plt
import numpy as np

# Path ke gambar asli
img_path = '../Dataset_Citra_Pesawat/AIRBUS A320/AIRBUS A320 - (7).jpg'

# Load gambar asli
img = load_img(img_path)
img_array = img_to_array(img)
x = np.expand_dims(img_array, axis=0)

# Fungsi bantu untuk visualisasi 1 augmentasi
def show_augmented_image(datagen, title):
    aug_iter = datagen.flow(x, batch_size=1)
    augmented = next(aug_iter)[0].astype(np.uint8)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Tampilkan gambar asli
plt.imshow(img_array.astype('uint8'))
plt.title("Gambar Asli")
plt.axis('off')
plt.show()

# Ilustrasi masing-masing teknik augmentasi
show_augmented_image(
    ImageDataGenerator(rotation_range=20),
    "Ilustrasi Data Augmentation - rotation_range=20"
)

show_augmented_image(
    ImageDataGenerator(width_shift_range=0.2),
    "Ilustrasi Data Augmentation - width_shift_range=0.2"
)

show_augmented_image(
    ImageDataGenerator(height_shift_range=0.2),
    "Ilustrasi Data Augmentation - height_shift_range=0.2"
)

show_augmented_image(
    ImageDataGenerator(horizontal_flip=True),
    "Ilustrasi Data Augmentation - horizontal_flip=True"
)
```
