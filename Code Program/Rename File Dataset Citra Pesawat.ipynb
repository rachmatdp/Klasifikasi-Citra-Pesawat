```python
import os

# Path ke folder utama dataset
dataset_dir = "../Dataset_Citra_Pesawat"  # contoh: './dataset'

for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)

    if os.path.isdir(class_path):
        files = sorted([
            f for f in os.listdir(class_path)
            if f.lower().endswith(('.jpg', '.jpeg'))
        ])
        
        success = 0
        failed = []

        for idx, filename in enumerate(files, start=1):
            new_name = f'{class_name} - ({idx}).jpg'  # Ubah semua ke .jpg
            old_path = os.path.join(class_path, filename)
            new_path = os.path.join(class_path, new_name)

            try:
                if not os.path.exists(new_path):  # Hindari menimpa file
                    os.rename(old_path, new_path)
                    success += 1
                else:
                    failed.append((filename, "Nama tujuan sudah ada"))
            except Exception as e:
                failed.append((filename, str(e)))

        print(f'{class_name}: {success}/{len(files)} file berhasil di-rename.')

        if failed:
            print("  ➤ File gagal di-rename:")
            for fname, reason in failed:
                print(f"    - {fname} → {reason}")
```

```python
import os
import re

# Path ke folder utama dataset
dataset_dir = "../Dataset_Citra_Pesawat"

for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)

    if os.path.isdir(class_path):
        # Ambil semua file gambar (jpg/jpeg)
        files = sorted([
            f for f in os.listdir(class_path)
            if f.lower().endswith(('.jpg', '.jpeg'))
        ])

        # Cari nomor terbesar yang sudah dipakai sebelumnya
        existing_numbers = []
        pattern = re.compile(rf"^{re.escape(class_name)} - \((\d+)\)\.jpg$", re.IGNORECASE)

        for f in os.listdir(class_path):
            match = pattern.match(f)
            if match:
                existing_numbers.append(int(match.group(1)))

        next_idx = max(existing_numbers) + 1 if existing_numbers else 1

        success = 0
        failed = []

        for filename in files:
            # Lewati jika sudah sesuai format nama tujuan
            if pattern.match(filename):
                continue

            new_name = f'{class_name} - ({next_idx}).jpg'
            old_path = os.path.join(class_path, filename)
            new_path = os.path.join(class_path, new_name)

            try:
                if not os.path.exists(new_path):
                    os.rename(old_path, new_path)
                    success += 1
                    next_idx += 1
                else:
                    failed.append((filename, "Nama tujuan sudah ada"))
            except Exception as e:
                failed.append((filename, str(e)))

        print(f'{class_name}: {success}/{len(files)} file berhasil di-rename.')

        if failed:
            print("  ➤ File gagal di-rename:")
            for fname, reason in failed:
                print(f"    - {fname} → {reason}")
```

```python
import os

# Path ke folder utama dataset
dataset_dir = "../Dataset_Citra_Pesawat"  # contoh: './dataset'

for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)

    if os.path.isdir(class_path):
        # Ambil semua file .jpg saja
        files = sorted([
            f for f in os.listdir(class_path)
            if f.lower().endswith('.jpg')
        ])

        success = 0
        failed = []

        # Step 1: Rename sementara untuk hindari konflik
        temp_names = []
        for idx, filename in enumerate(files):
            old_path = os.path.join(class_path, filename)
            temp_name = f'temp_{idx}.tmp'
            temp_path = os.path.join(class_path, temp_name)
            os.rename(old_path, temp_path)
            temp_names.append(temp_name)

        # Step 2: Rename ke nama final terurut
        for idx, temp_name in enumerate(temp_names, start=1):
            new_name = f'{class_name} - ({idx}).jpg'
            temp_path = os.path.join(class_path, temp_name)
            new_path = os.path.join(class_path, new_name)

            try:
                os.rename(temp_path, new_path)
                success += 1
            except Exception as e:
                failed.append((temp_name, str(e)))

        print(f'{class_name}: {success}/{len(files)} file berhasil di-rename ulang.')

        if failed:
            print("  ➤ File gagal di-rename:")
            for fname, reason in failed:
                print(f"    - {fname} → {reason}")
```

```python
dataset_path = "../Dataset_Citra_Pesawat"
dataset_info = {}

for label in os.listdir(dataset_path):
    label_path = os.path.join(dataset_path, label)
    if os.path.isdir(label_path):
        file_count = len(os.listdir(label_path))
        dataset_info[label] = file_count

for label, count in dataset_info.items():
    print(f"Label: {label}, Jumlah Citra: {count}")
    
total_images = sum(dataset_info.values())
print(f"Total Jumlah Citra: {total_images}")
```
